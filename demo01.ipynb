```python
class NhanVien:
    def __init__(self, ma_nv, ho_ten, ngay_sinh, dia_chi, dien_thoai):
        self.ma_nv = ma_nv
        self.ho_ten = ho_ten
        self.ngay_sinh = ngay_sinh
        self.dia_chi = dia_chi
        self.dien_thoai = dien_thoai

    def hien_thi_thong_tin(self):
        print("Mã NV:", self.ma_nv)
        print("Họ tên:", self.ho_ten)
        print("Ngày sinh:", self.ngay_sinh)
        print("Địa chỉ:", self.dia_chi)
        print("Điện thoại:", self.dien_thoai)


class NhanVienHanhChinh(NhanVien):
    def __init__(self, ma_nv, ho_ten, ngay_sinh, dia_chi, dien_thoai, he_so_luong, luong_co_ban):
        super().__init__(ma_nv, ho_ten, ngay_sinh, dia_chi, dien_thoai)
        self.he_so_luong = he_so_luong
        self.luong_co_ban = luong_co_ban

    def tinh_luong(self):
        luong = self.he_so_luong * self.luong_co_ban
        return max(luong, 4000000)

    def hien_thi_thong_tin(self):
        super().hien_thi_thong_tin()
        print("Lương:", self.tinh_luong())


class GiangVien(NhanVien):
    def __init__(self, ma_nv, ho_ten, ngay_sinh, dia_chi, dien_thoai, he_so_chuc_danh, so_tiet_giang, don_gia_tiet_giang):
        super().__init__(ma_nv, ho_ten, ngay_sinh, dia_chi, dien_thoai)
        self.he_so_chuc_danh = he_so_chuc_danh
        self.so_tiet_giang = so_tiet_giang
        self.don_gia_tiet_giang = don_gia_tiet_giang

    def tinh_luong(self):
        return self.he_so_chuc_danh * self.so_tiet_giang * self.don_gia_tiet_giang

    def hien_thi_thong_tin(self):
        super().hien_thi_thong_tin()
        print("Lương:", self.tinh_luong())


class QuanLyNhanVien:
    def __init__(self):
        self.danh_sach_nhan_vien = []

    def them_nhan_vien(self, nhan_vien):
        self.danh_sach_nhan_vien.append(nhan_vien)

    def xoa_nhan_vien(self, ma_nv):
        for nv in self.danh_sach_nhan_vien:
            if nv.ma_nv == ma_nv:
                self.danh_sach_nhan_vien.remove(nv)
                print("Đã xoá nhân viên có mã:", ma_nv)
                return
        print("Không tìm thấy nhân viên có mã:", ma_nv)

    def thiet_lap_luong_co_ban(self, ma_nv, luong_co_ban):
        for nv in self.danh_sach_nhan_vien:
            if isinstance(nv, NhanVienHanhChinh) and nv.ma_nv == ma_nv:
                nv.luong_co_ban = luong_co_ban
                print("Đã thiết lập lương cơ bản cho nhân viên", ma_nv)
                return
        print("Không tìm thấy nhân viên hành chính có mã:", ma_nv)

    def thiet_lap_don_gia_tiet_giang(self, ma_nv, don_gia_tiet_giang):
        for nv in self.danh_sach_nhan_vien:
            if isinstance(nv, GiangVien) and nv.ma_nv == ma_nv:
                nv.don_gia_tiet_giang = don_gia_tiet_giang
                print("Đã thiết lập đơn giá tiết giảng cho giảng viên", ma_nv)
                return
        print("Không tìm thấy giảng viên có mã:", ma_nv)

    def tinh_muc_luong_trung_binh(self):
        tong_luong = 0
        so_luong = 0
        for nv in self.danh_sach_nhan_vien:
            if isinstance(nv, NhanVienHanhChinh):
                tong_luong += nv.tinh_luong()
                so_luong += 1
        if so_luong == 0:
            return 0
        return tong_luong / so_luong

    def hien_thi_nhan_vien_luong_thap_hon_trung_binh(self):
        muc_luong_trung_binh = self.tinh_muc_luong_trung_binh()
        print("Mức lương trung bình trường:", muc_luong_trung_binh)
        print("Danh sách nhân viên có mức lương thấp hơn mức lương trung bình trường:")
        for nv in self.danh_sach_nhan_vien:
            if isinstance(nv, NhanVienHanhChinh) and nv.tinh_luong() < muc_luong_trung_binh:
                nv.hien_thi_thong_tin()
            elif isinstance(nv, GiangVien) and nv.tinh_luong() < muc_luong_trung_binh:
                nv.hien_thi_thong_tin()


# Test chương trình
quan_ly_nv = QuanLyNhanVien()

# Thêm nhân viên hành chính
nhan_vien_hanh_chinh1 = NhanVienHanhChinh("NVHC001", "Nguyen Van A", "01/01/1990", "TP.HCM", "123456789", 2.0, 5000000)
nhan_vien_hanh_chinh2 = NhanVienHanhChinh("NVHC002", "Tran Thi B", "02/02/1991", "TP.HCM", "987654321", 1.8, 4500000)
quan_ly_nv.them_nhan_vien(nhan_vien_hanh_chinh1)
quan_ly_nv.them_nhan_vien(nhan_vien_hanh_chinh2)

# Thêm giảng viên
giang_vien1 = GiangVien("GV001", "Le Van C", "03/03/1980", "TP.HCM", "135792468", 3.5, 40, 300000)
giang_vien2 = GiangVien("GV002", "Pham Thi D", "04/04/1975", "TP.HCM", "246813579", 4.0, 30, 350000)
quan_ly_nv.them_nhan_vien(giang_vien1)
quan_ly_nv.them_nhan_vien(giang_vien2)

# Thiết lập lương cơ bản cho nhân viên hành chính và đơn giá tiết giảng cho giảng viên
quan_ly_nv.thiet_lap_luong_co_ban("NVHC001", 5500000)
quan_ly_nv.thiet_lap_luong_co_ban("NVHC002", 5000000)
quan_ly_nv.thiet_lap_don_gia_tiet_giang("GV001", 320000)
quan_ly_nv.thiet_lap_don_gia_tiet_giang("GV002", 360000)

# Hiển thị danh sách nhân viên có mức lương thấp hơn mức lương trung bình trường
quan_ly_nv.hien_thi_nhan_vien_luong_thap_hon_trung_binh()
```
